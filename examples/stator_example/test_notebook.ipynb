```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
output_path = 'output_test_3'

stator_data = pd.read_csv('./' + output_path + '/stator.csv')
welding_data = pd.read_csv('./' + output_path + '/welding.csv')
```

```python
stator_data.loc[stator_data["station_id"] == 2]
```

```python
def plot_resistance():

     surface_data = pd.read_csv('./' + output_path + '/stator.csv')

     labels = ['assemble_stator', 'dry_glue', 'weld_connections']

     plt.figure(figsize=(25,7))
     plt.plot(surface_data[surface_data['station_id'] == 2]['time'],
              surface_data[surface_data['station_id'] == 2]['el_resistance_mOhm'], label="label")

     plt.xlabel('Sim. time')
     plt.ylabel('Resistance in mOhm')
     plt.title('Resistance over simulated time')
     plt.legend(loc='upper right')

     # plt.savefig('./' + output_path + '/shaft_surface.png')
```

```python
plot_resistance()
```
